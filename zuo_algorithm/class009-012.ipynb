#### 009 单双链表及其反转
**值传递与引用传递（python视角）**
在 Python 中，所有变量都是对象的引用，但对象的可变性决定了函数内修改是否影响外部：
- 不可变对象（int、float、str、tuple 等）：函数内对参数重新赋值（如 a = 0）只是将局部变量指向新对象，外部引用不受影响，行为类似于值传递。

- 可变对象（list、dict、自定义类实例等）：函数内修改对象内容（如 b.val = 6 或 c[0] = 100）会影响外部引用指向的同一对象；但将局部变量重新赋值为 None 不会影响外部引用。

```python
# 导入必要的模块（本示例不需要额外导入）

class ListReverse:
    """
    演示值传递与引用传递，并提供单链表和双链表的反转算法。
    对应 Java 原版 class009.ListReverse。
    """

    @staticmethod
    def main():
        """
        主函数：展示基本类型和对象类型在函数调用时的行为差异。
        """
        # ---------- 不可变类型（类似于值传递） ----------
        a = 10                     # 整数是不可变对象
        ListReverse.f(a)           # 传递引用，但整数不可变，函数内重新赋值不影响外部
        print(a)                   # 输出 10

        # ---------- 可变对象（类似于引用传递） ----------
        b = ListReverse.Number(5)  # 创建自定义对象
        ListReverse.g1(b)          # 函数内将局部变量指向 None，不影响外部引用
        print(b.val)               # 输出 5
        ListReverse.g2(b)          # 函数内修改对象的属性
        print(b.val)               # 输出 6

        # ---------- 列表（可变对象） ----------
        c = [1, 2, 3, 4]           # 列表是可变对象
        ListReverse.g3(c)          # 函数内将局部变量指向 None，不影响外部引用
        print(c[0])                # 输出 1
        ListReverse.g4(c)          # 函数内修改列表元素
        print(c[0])                # 输出 100

    @staticmethod
    def f(a):
        """尝试修改整数参数（不可变对象）"""
        a = 0                      # 局部变量 a 重新绑定到新对象，外部不受影响

    class Number:
        """简单的包装类，包含一个整数属性 val"""
        def __init__(self, v):
            self.val = v

    @staticmethod
    def g1(b):
        """将局部引用设为 None，不影响外部引用"""
        b = None

    @staticmethod
    def g2(b):
        """修改对象的属性，影响外部对象"""
        b.val = 6

    @staticmethod
    def g3(c):
        """将局部引用设为 None，不影响外部列表"""
        c = None

    @staticmethod
    def g4(c):
        """修改列表元素，影响外部列表"""
        c[0] = 100

    # -------------------- 单链表节点定义 --------------------
    class ListNode:
        """
        单链表节点类, 
        有一个节点结构Node, Node有一个自己的值 val, 指向下一个节点的指针 next
        """
        def __init__(self, val, next=None):
            self.val = val
            self.next = next

    # -------------------- 单链表反转 --------------------
    class Solution:
        """对应原 Java 的内部类 Solution，提供反转单链表方法"""
        @staticmethod
        def reverseList(head):
            """
            反转单链表
            :param head: 原链表头节点
            :return: 反转后的新头节点
            """
            pre = None          # 前驱节点，初始为 None
            nxt = None          # 后继节点占位
            while head is not None:
                #节点从头节点head开始，这两步相当于把当前的节点的指针反转
                nxt = head.next      # 保存下一个节点
                head.next = pre      # 反转当前节点的 next 指针

                #当前节点的指针反转完成，接下来要移动到下一个节点继续进行反转
                pre = head           # 前驱后移, 即原本pre在head的前一位，现在后移到了head的位置
                head = nxt           # 当前节点后移, head后移到head.next的位置
            return pre               # 新的头节点

    # -------------------- 双链表节点定义 --------------------
    class DoubleListNode:
        """
        双链表节点类,
        有一个节点结构Node, Node有一个自己的值 val, 指向下一个节点的指针 next, 指向上一个节点的指针 last
        """
        def __init__(self, v):
            self.value = v
            self.last = None    # 前驱指针
            self.next = None    # 后继指针

    # -------------------- 双链表反转 --------------------
    @staticmethod
    def reverseDoubleList(head):
        """
        反转双链表
        :param head: 原双链表头节点
        :return: 反转后的新头节点
        """
        pre = None
        nxt = None
        while head is not None:
            # 这三步相当于把当前的节点的指针反转
            nxt = head.next          # 保存原后继
            head.next = pre          # 反转 next 指针
            head.last = nxt          # 反转 last 指针（原后继变成前驱）

            # 当前节点的指针反转完成，接下来要移动到下一个节点继续进行反转
            pre = head               # 前驱后移
            head = nxt               # 当前节点后移
        return pre                   # 新头节点


# 程序入口
if __name__ == "__main__":
    ListReverse.main()
```

#### 合并两个有序列表

```python
class ListNode:
    """
    单链表节点类
    对应 Java 中的 ListNode
    """
    def __init__(self, val, next=None):
        """
        构造函数，支持只传 val 或同时传 val 和 next
        :param val: 节点值
        :param next: 下一个节点，默认为 None
        """
        self.val = val
        self.next = next


class Solution:
    """
    合并两个有序链表的解决方案类
    对应 Java 中的内部类 Solution
    """

    @staticmethod
    def mergeTwoLists(head1, head2):
        """
        合并两个升序链表，返回合并后的升序链表头节点
        算法使用原地拼接，不创建新节点

        :param head1: 第一个升序链表的头节点
        :param head2: 第二个升序链表的头节点
        :return: 合并后链表的头节点
        """
        # 处理空链表情况：如果任一链表为空，直接返回另一个
        if head1 is None or head2 is None:
            # 条件表达式：若 head1 为空则返回 head2，否则返回 head1
            return head2 if head1 is None else head1

        # 确定合并后的头节点：取两个头节点中值较小的那个
        head = head1 if head1.val <= head2.val else head2

        # cur1 指向较小头节点的下一个节点（即该链表的剩余部分）
        cur1 = head.next
        # cur2 指向另一个链表的头节点
        cur2 = head2 if head is head1 else head1

        # pre 始终指向已合并部分的最后一个节点
        pre = head

        # 同时遍历两个链表，每次选择值较小的节点接到 pre 后面
        while cur1 is not None and cur2 is not None:
            if cur1.val <= cur2.val:
                pre.next = cur1      # 将 cur1 接入合并链表
                cur1 = cur1.next     # cur1 后移
            else:
                pre.next = cur2      # 将 cur2 接入合并链表
                cur2 = cur2.next     # cur2 后移
            pre = pre.next            # pre 后移到新加入的节点

        # 将剩余未遍历完的链表直接接在后面
        pre.next = cur1 if cur1 is not None else cur2

        return head
```

#### 链表相加

```python
'''哑节点(Dummy Node): 本质上是一个额外的、不存储有效数据的节点，通常作为链表头部的占位符。这个节点本身并不属于最终结果的一部分，但它的 next 指针会指向真正的第一个有效节点。'''

class ListNode:
    '''单链表定义类'''
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        """
        将两个逆序存储的链表表示的整数相加，返回结果链表的头节点
        :param l1: 第一个链表头节点
        :param l2: 第二个链表头节点
        :return: 结果链表头节点
        """
        # 哑节点，简化链表构建过程
        dummy = ListNode(0)   # 初始化一个哑节点作为结果链表的虚拟头，其 next 最终指向真正的头节点
        cur = dummy    # 指向当前链表的最后一个节点
        carry = 0  # 进位

        # 同时遍历两个链表，直到两者都为空
        while l1 or l2 or carry:
            # 获取当前位的值（如果链表已结束则取0）
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0

            # 计算当前位的和以及新的进位
            total = v1 + v2 + carry
            carry = total // 10
            val = total % 10

            # 创建新节点并移动cur指针
            cur.next = ListNode(val)
            cur = cur.next

            # 移动链表指针（如果存在）
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next    # 哑节点的 next 指针会指向真正的第一个有效节点。
```

#### 划分链表

```python
class ListNode:
    """单链表节点类"""
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def partition(self, head: ListNode, x: int) -> ListNode:
        '''
        将链表分割为小于x和大于等于x的两部分, 并且不改变每一部分的相对顺序
        head: 原链表的头节点
        x: 分隔值
        return: 分隔后的链表的头节点
        '''

        # 初始化分隔值左右两部分链表的头尾指针
        left_head = None
        left_tail = None
        right_head = None
        right_tail = None

        # 遍历链表
        while head:
            nxt = head.next
            head.next = None    # 将当前节点与原链表断开

            if head.val < x:
                if left_head is None:
                    left_head = head
                else:
                    left_tail.next = head    # 把当前节点移到左区域的最后一个节点

                left_tail = head    # 更新左区域尾指针

            else:
                if right_head is None:
                    right_head = head
                else:
                    right_tail.next = head

                right_tail = head

            head = nxt

        # 如果左区域为空，直接返回右区域
        if left_head is None:
            return right_head

        # 否则将左右区域连接起来
        left_tail.next = right_head
        return left_head
```
