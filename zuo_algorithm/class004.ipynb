### Class004 三傻排序


```python
class SelectBubbleInsert():
    '''选择排序、冒泡排序、插入排序的实现类'''

    @staticmethod
    def swap(arr, i, j):
        '''交换数组中两个元素的位置'''
        arr[i], arr[j] = arr[j], arr[i]

    @staticmethod
    def select(arr):
        '''选择排序: 数组长度为N, 从0到N-1位置,找到并选择最小值与位置0交换,然后从1到N-1位置,找到并选择最小值与位置1交换,依次类推'''
        
        # 如果数组为空或数组长度小于2，则不需要排序
        if arr is None or len(arr) < 2:
            return

        n = len(arr)
        # 外层循环，遍历数组，i从0到n-2
        for i in range(n-1):
            # 内层循环，当前范围为从i到n-1, 便利当前范围, 找到其中的最小值，然后交换到最前面
            min_index = i
            for j in range(i+1, n):
                if arr[j] < arr[min_index]:    # 如果当前元素小于最小值，则更新最小值索引
                    min_index = j

            if min_index != i:    # 如果最小值索引不是i，则交换最小值和i位置的元素
                SelectBubbleInsert.swap(arr, i, min_index)

    @staticmethod
    def bubble(arr):
        '''冒泡排序: 数组长度为N, 从位置0到位置N-1, 依次比较相邻两元素,较大者滑动到靠前的位置, 最后范围0到N-1中的最大者会到位置N-1,然后范围0到N-2,依次类推'''

        # 如果数组为空或数组长度小于2，则不需要排序
        if arr is None or len(arr) < 2:
            return

        n = len(arr)
        # 外层循环，end从n-1到1
        for end in range(n-1, 0, -1):    # 步长-1, 表示从最后一个位置n-1倒着遍历到位置1
            # 内层循环，从0到end-1, 依次比较相邻两元素,较大者滑动到靠前的位置
            for i in range(end):
                if arr[i] > arr[i+1]:
                    SelectBubbleInsert.swap(arr, i, i+1)

    @staticmethod
    def insert(arr):
        '''插入排序: 前面序列已经有序, 将新元素从右到左依次滑动, 滑动至没有更小值为止'''

        # 如果数组为空或数组长度小于2，则不需要排序
        if arr is None or len(arr) < 2:
            return

        n = len(arr)
        # 外层循环，第一个元素形成的序列天然有序，因此插入的新元素的范围是1到n-1
        for i in range(1, n):
            j = i-1
            # 内层循环，新元素与前面有序序列中的元素依次比较,
            while j >= 0 and arr[j] > arr[j+1]:
                SelectBubbleInsert.swap(arr, j, j+1)
                j -= 1

def test():
    test_arr = [
        [64, 34, 25, 12, 22, 11, 90],  # 随机数组
        [1, 2, 3, 4, 5, 6, 7],  # 已排序数组
        [7, 6, 5, 4, 3, 2, 1],  # 逆序数组
        [5],  # 单元素数组
        [],  # 空数组
        [3, 1, 4, 1, 5, 9, 2, 6, 5]  # 包含重复元素的数组
    ]

    algorithms = [
        ('选择排序', SelectBubbleInsert.select),
        ('冒泡排序', SelectBubbleInsert.bubble),
        ('插入排序', SelectBubbleInsert.insert)
    ]

    for algo_name, algo_func in algorithms:
        print(f"测试{algo_name}排序:")

        for i, arr in enumerate(test_arr):
            
            arr_copy = arr.copy()
            arr_origin = arr.copy()

            print(f"测试第{i+1}个数组: {arr_origin}")

            algo_func(arr_copy)
            expected = sorted(arr_origin)
            if arr_copy == expected:
                print(f"第{i+1}个数组排序成功, 排序结果: {arr_copy}")
            else:
                print(f"第{i+1}个数组排序失败")
                print(f"排序结果: {arr_copy}")
                print(f"预期结果: {expected}")

if __name__ == "__main__":
    test()
```
