### 005 对数器
**非常重要的自我验证技巧**
构造一个暴力解、构造一个最优解(想要测试的方法)，随机生成一个随机样本数据，在暴力解和最优解下分别得到答案，如果很多次下的随机数据在在暴力解和最优解下分别得到的答案都相同，能够对上，说明最优解的想法是正确的。
如果答案对不上，就打印出来出错的随机样本，用这个出错样本来带入代码一步一步排查(print大法或断点技术)，改对暴力解和最优解

### 006 二分搜索入门

```python
# 二分搜索: 在有序数组中寻找是否存在一个数
import random
import sys

def main():
    '''对数器测试函数'''
    N = 100    # 随机数组最大长度
    V = 1000    # 随机数组中每个数的取值范围
    test_time = 50000    # 测试次数

    print('测试开始')

    for i in range(test_time):
        # 随机初始化测试数组的长度，长度范围为[0, N-1]
        n = int(random.random() * N)    

        # 生成随机数组并进行排序
        arr = random_array(n, V)
        arr.sort()    # 列表中的sort()方法，原地对数组进行排序

        # 随机生成要查找的数字
        num = int(random.random() * V)

        if right(arr, num) != exist(arr, num):
            print("出错了")
            break

    print("测试结束")

def random_array(n, v):
    '''生成随机数组
    n: 数组长度
    v: 数组中每个数字的取值范围(1到v)
    返回包含n个随机整数的列表
    
    random.random(): 得到[0,1)之间的一个随机小数
    random.random() * v: 得到[0,v)之间的一个随机小数
    int(random.random() * v): 得到[0,v-1]之间的一个随机整数
    int(random.random() * v) + 1: 得到[1,v]之间的一个随机整数
    '''
    return [int(random.random() * v) + 1 for _ in range(n)]


def right(arr, num):
    '''用来进行对数器测试的暴力验证方法'''
    if arr is None or len(arr) == 0:
        return False
    
    for i in range(len(arr)):
        if arr[i] == num:
            return True

    return False

def exist(arr, num):
    '''用来验证的自己的二分搜索优化算法
    arr: 已排序数组
    num: 要查找的目标数字
    '''
    if arr is None or len(arr) == 0:
        return False

    l = 0    # 左指针初始化为0
    r = len(arr) - 1    # 右指针初始化为数组最右边的数
    m = 0    # 中间指针初始化为0

    while l <= r:
        m = l + ((r - l) >> 1)    # 中间指针指向数组的中间位置
        if arr[m] == num:
            return True
        elif arr[m] > num:    # 如果中间位置 m 的值大于要找的数，则右边界更新为 m-1
            r = m -1
        elif arr[m] < num:    # 如果中间位置 m 的值小于要找的数，则左边界更新为 m+1
            l = m + 1

    return False    # 循环结束都没有return True, 表示还没有找到，此时返回False

if __name__ == "__main__":
    main()
```

```python
# 二分数组: 寻找有序数组中 ≤ num 的最右边的数
def random_array6_2(n, v):
    '''
    生成随机数组
    n: 数组长度
    v: 数组中每个数的取值范围, 1到v
    返回一个随机数组
    '''
    return [int(random.random() * v) + 1 for _ in range(n)]

def right6_2(arr, num):
    '''
    暴力解法
    '''
    if arr is None or len(arr) == 0:
        return -1

    # for i in range(len(arr)-1, -1, -1):    # 从右往左遍历
    #     if arr[i] <= num:
    #         return i
    
    i = 0
    while i <= len(arr) - 1:
        if arr[i] <= num:
            i += 1
        else:
            return i - 1

    return len(arr) - 1    # 循环结束，说明所有元素都 ≤ num

def exist6_2(arr, num):
    '''
    要验证的二分搜索的算法
    '''
    if arr is None or len(arr) == 0:
        return -1

    l = 0    # 左指针初始为0
    r = len(arr) - 1    # 右指针
    ans = -1    # 答案索引初始为-1

    while l <= r:
        m = l + ((r - l) >> 2)    # 中间位置初始化索引
        if arr[m] <= num:    # 如果中间位置值小于等于目标数num, 则更新ans和左边界
            ans = m
            l += 1
        else:    # 如果中间位置值大于目标数num, 则不更新ans，更新左边界
            r -= 1

    return ans

def main():
    '''对数器主要测试函数'''
    N = 5    # 数组最大长度
    V = 10    # 数组中每个数的范围
    test = 50    # 测试次数
    print('测试开始')

    for i in range(test):
        # 随机确定数组的长度值
        n = int(random.random() * N)    

        # 生成随机数组并排序
        arr = random_array6_2(n, V)
        arr.sort()
        #print(arr)

        # 随机生成一个待查找的数字num
        num = int(random.random() * V)
        #print(num)

        if right6_2(arr, num) != exist6_2(arr, num):
            print('出错了')
            print(f"arr={arr}, num={num}, right={right6_2(arr, num)}, exist={exist6_2(arr, num)}")

    print('测试结束')

if __name__ == "__main__":
    main()
```

```python
# 二分搜索: 峰值问题
class Solution:
    '''
    寻找峰值元素的类
    '''

    @staticmethod    # 声明静态方法，可通过类名直接调用, 不依赖于类或实例的状态，也就是说它不能访问实例属性（self）或类属性（cls），它就像一个定义在类内部的普通函数，只是通过类名来调用
    def findPeak(arr):
        """
        在数组 arr 中寻找任意一个峰值元素的索引
        时间复杂度 O(log n)，空间复杂度 O(1)

        参数:
            arr: List[int] -- 整数数组，已知任何相邻元素不相等

        返回:
            int -- 峰值元素的索引
        """
        n = len(arr)

        # 如果数组只有一个元素，该元素即为峰值
        if n == 1:                        
            return 0                       # 返回索引0

        # 检查第一个元素和最后一个元素是否为峰值
        if arr[0] > arr[1]:                # 通过索引访问列表元素，> 比较运算符
            return 0

        if arr[n - 1] > arr[n - 2]:        # 索引 n-1 是最后一个元素
            return n - 1

        # 第一个元素和最后一个元素均不是峰值，在中间部分进行二分搜索查找峰值
        l = 1                              # 左边界，从第二个元素开始
        r = n - 2                          # 右边界，到倒数第二个元素结束
        ans = -1                           # 答案初始化为 -1（未找到）

        while l <= r:                      # while 循环，当左边界 <= 右边界时继续
            # 计算中点索引，使用整数除法 // 确保结果为整数
            m = (l + r) // 2                # 等价于 Java 中的 (l + r) / 2（自动截断）

            # 如果中点左边的元素比中点大，说明峰值在左侧区域
            if arr[m - 1] > arr[m]:         # 比较相邻元素
                r = m - 1                    # 缩小右边界到 m-1

            # 如果中点右边的元素比中点大，说明峰值在右侧区域
            elif arr[m] < arr[m + 1]:        # elif 是 else if 的缩写
                l = m + 1                    # 增大左边界到 m+1

            # 否则，中点元素就是峰值（比左右都大）
            else:
                ans = m                      # 记录答案
                break                        # break 跳出循环

        return ans                           # 返回峰值索引
```
